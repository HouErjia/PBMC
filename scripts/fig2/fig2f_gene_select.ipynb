```r
# 读取 CSV 文件
data <- read.csv('/home/houerjia/program/experments/PBMC_handoff/PBMC_sqlou/fig2/fig2c_translation_efficiency_changes_with_gene_info_byH.csv')

# 定义要筛选的基因列表
genes_to_select <- c('TNFSF13', 'CCL25', 'CD93', 'CD82', 'CD24', 'RELT')

# 筛选出指定基因的数据
selected_data <- data[data$external_gene_name %in% genes_to_select, ]

# 查看筛选后的数据
selected_data
```

```r
# 选择指定列组成新表
new_data <- selected_data[, c("external_gene_name","ensembl_gene_id","TE_H", "TE_W")]

# 输出新表
new_data
```

```r
# 使用 colnames() 修改列名
colnames(new_data)[colnames(new_data) == 'TE_H'] <-'H'
colnames(new_data)[colnames(new_data) == 'TE_W'] <-'W'
colnames(new_data)[colnames(new_data) == 'external_gene_name'] <- ''
colnames(new_data)[colnames(new_data) == 'ensembl_gene_id'] <- ''

colnames(new_data)
```

```r
write.csv(new_data, "/home/houerjia/program/experments/PBMC_handoff/PBMC_sqlou/fig2/fig2f_selected_genes_data_byH.csv", row.names = FALSE)
```
